```python
# iloc -> integer location

# CSV
poke_list = pd.read_csv('pokemon_data.csv')

# XLSX
# poke_list_xl = pd.read_excel('pokemon_data.xlsx')

# TEXT FILE
# poke_list_txt = pd.read_csv('pokemon_data.txt', delimiter='\t') # the text file are tab spereted thef we need \t as delimeter , for any other formatting for gap between cols are passed in as delimeter

# print(poke_list)

# read top rows or bottom rows

# print(poke_list.head(3))
# print(poke_list.tail(3))
```

# READING

```python
# HEADERS
print(poke_list.columns)

# READ EACH COLUMN
print(poke_list['Name'])
print(poke_list['Name'])
print(poke_list.Name)
print(poke_list['Name'][0:5])

# MULTIPLE COLUMNS
print(poke_list[['Name', 'HP', 'Type 1']])

# READ EACH ROW
print(poke_list.head())
print(poke_list.iloc[0:4])

# ITERATION OVER ROWS
for index, rows in poke_list.iterrows():
  print(index, rows['Name'])


# FILTERING

# list.loc[list[value] ->condition,  "Value"]
# above specifies that the first parameter is independent of the list.loc bcz it is a condition hence in list we find the location of the element(specified as second arguement) when the condition is met
# outcome -> condn? -> poke_list.loc["Fire"]
poke_list.loc[poke_list["Type 1"] == "Fire"]

# READ A SPECIFIC VALUE
print(poke_list.iloc[2,1])
```

SORTING/DESCRIBING DATA

```python
# describe gives all the high level data like mean and std deviation type stats

# poke_list.describe()

# poke_list.sort_values(['Name'] , ascending=False)

#by default both will be sort in ascending order
poke_list.sort_values(["Type 1" ,"HP"], ascending=[1,0]) #HP is in descending order and Type1 is in ascending order

# axis=0 -> vertically , axis=0 -< horizontally
```

# MAKING CHANGES TO DATA

```python
# BASIC WAY

poke_list['Total'] = poke_list['HP'] + poke_list['Attack'] + poke_list['Defense'] + poke_list['Sp. Atk'] + poke_list['Sp. Def'] + poke_list['Speed']
print(poke_list)

# DROP A COLUMN

# n.drop() return the data which is need to be stored in the original data
poke_list = poke_list.drop(columns=['Total'])
print(poke_list)

# ANOTHER WAY
poke_list['Total'] = poke_list.iloc[: ,4:10].sum(axis=1)
print(poke_list)

# RE-ORDERING
yo = poke_list.columns.values
cols = list(poke_list.columns)
print(cols[0:4]) #gives from '#' to 'Type 2' return as list bcz of range
print(cols[-1]) #gives total as its last element , but its single so 'Total'
print(cols[4:12])

poke_list = poke_list[cols[0:4] + [cols[-1]] + cols[4:12]]
print(poke_list)
```

# FILTERING

```python
filtered_list = poke_list.loc[(poke_list["Type 1"] == "Grass") & (poke_list["Type 2"] == "Poison") & (poke_list["HP"] > 70)]

# we see above, filtered_list return indices that are filtered , to create new indices for filtered list we use reset_index , but the original indices aren't deleted to delete them we specify drop=True
filtered_list = filtered_list.reset_index(drop=True)
print(filtered_list)
```

# Working with large amount of data

```python
# For large dataset we don't read whole data at ones we read it in form of chunks, chunks size specify at which event the data is called for again

# for df in pd.read_csv('pokemon_data.csv', chunksize=5):
#   print('Print row as chunk of 5')
#   print(df)


# use case

new_df = pd.DataFrame(columns = poke_list.columns)

for df in pd.read_csv('pokemon_data.csv', chunksize=5):
  res = poke_list.groupby("Type 1").count()
  new_df = pd.concat([new_df, res])

print(new_df)

# usefull in case of large data available and we are performing some operations on it and finally store the result of the oprs in the new_df this reduce the size of data by lot
```
